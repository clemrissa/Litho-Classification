# Lithology Labelling
To help you during data preparation. This code allows you to take specified log interval with known lithology. It will help you a lot if you work with raw data and  a lot of wells.

```python
import lasio
import pandas as pd
import numpy as np
import math
import sys
np.set_printoptions(threshold=sys.maxsize)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

dset = pd.DataFrame()
```

Insert the file path, top and bottom interval, Depth Driller (DD), Depth Logger (DL), the datatype, and Lithology. Modify the step as necessary. I worked with LAS file with null step and sometimes inverted depth in this case, not a good standard.

```python
l = lasio.read("your file path")
top = 2370
bottom = 2652
DD = 3142
DL = 3142
offset = DL-DD
dataType = 'Log'
Lithology = ' Andesite'
well = l.well.WELL.value
start = l.well.STRT.value
stop = l.well.STOP.value
step = l.well.STEP.value
df = l.df()

def convertDepth(top,bottom,offset,start,stop,step):
    if step == 0:
        step = 0.1524
    if start > stop:
        a = start
        start = stop
        stop = a
    top = math.ceil((top+offset-start)/step)
    bottom = math.ceil((bottom+offset-start)/step)
    return top, bottom

top,bottom = convertDepth(top,bottom,offset,start,stop,step)
interval = pd.DataFrame(df.iloc[top:bottom].assign(DATATYPE=dataType))
interval.insert(0, 'WELL', well)
```

Run this if you want to assign more than one different interval in one well.

```python
dset = pd.concat([interval,dset],sort=False)
```

To export your data (one well) to textfile.

```python
print(dset.to_csv(r'%s .txt' %well, sep='\t', mode='a', na_rep='NULL'))
```
